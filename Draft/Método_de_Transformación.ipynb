### Manipulación de columnas

```python
df3 = pd.read_csv ("Data/datos_mergeados.csv")
df3.head()
```

Vamos establecer únicamente dos opciones para la columna AD_Situation: New, old. Para ello, hemos creado una nueva columna donde reflejar el resultado: "situación":

```python
df3["situación"] = df3.AD_Situation.str.extract("(\w+)", expand=True)
df3.head(2)
```

Vamos a cambiar la unidad de medida de "sqft" a "m2" sobre la columna "Measruing uni", y vamos a crear una columna nueva: "Bulding_space", donde vamos el Building space en m2. Para ello, vamos a dividir la columna "Building space	Measruing" por 10.764:

```python
df3['Measruing unit']= df3['Measruing unit'].replace(["sqft"],"m2")
df3['Building_space']= round(df3['Building space']/10.764, 0)
df3.head(2)
```

Vamos a eliminar la columna "State" y crear otra columna: "estado", únicamente mostrando los 3 Estado: California, Oregón, Washington:

```python
df3["estado"] = df3.State.str.extract("(\w+)", expand=True)
df3.head(2)
```

```python
df3['estado'] = df3['estado'].replace(["WA"], "Washington")
df3['estado'] = df3['estado'].replace(["OR"], "Oregón")
df3['estado'] = df3['estado'].replace(["CA"], "California")
df3.head(2)
```
